```python
from mqt.bench import get_benchmark
for i in all_algo:
    print("---------------",i,"---------------")
    qc_algorithmic_level = get_benchmark(benchmark_name= i, level="alg", circuit_size=3)
    print(qc_algorithmic_level)
    print("=================================================================================================")
```

<p><B>Method :2 </B>:: Calling at "Target-independent Level"</p>

```python
for i in all_algo:
    print("---------------",i,"---------------")
    qc = get_benchmark(benchmark_name=i, level="indep", circuit_size=3, compiler="qiskit")
    print(qc.draw())
    print("================================================================================================")
```

<p><B>Method :3 </B>:: Calling at "Target-dependent Native Gates Level" </p>

```python
from mqt.bench import CompilerSettings, QiskitSettings, TKETSettings, get_benchmark
compiler_settings = CompilerSettings(qiskit=QiskitSettings(optimization_level=1))
for i in all_algo:
    qc_native_gates_level = get_benchmark(
        benchmark_name=i,
        level="nativegates",
        circuit_size=3,
        compiler="qiskit",
        compiler_settings=compiler_settings,
        provider_name="ionq",)
    print("---------------",i,"---------------")
    print(qc_native_gates_level.draw())
    print("=================================================================================================")
```

<p><B>Method :4 </B>:: Calling at "Target-dependent Mapped Level"</p>

```python
from pytket.extensions.qiskit import tk_to_qiskit
compiler_settings = CompilerSettings(tket=TKETSettings(placement="lineplacement"))
for i in all_algo:
    qc_mapped_level = get_benchmark(
        benchmark_name=i,
        level="mapped",
        circuit_size=3,
        compiler="tket",
        compiler_settings=compiler_settings,
        device_name="ionq_harmony",)
    print("---------------",i,"---------------")
    print(tk_to_qiskit(qc_mapped_level).draw())
    print("=============================================================================================")
```
